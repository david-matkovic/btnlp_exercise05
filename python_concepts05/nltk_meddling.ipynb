What does this compound score mean? 
It's not necessarily taking the sum of pos, neu, neg values, right?

According to the VADER docs the compound score sums the valence scores of *each word* in the lexicon, and then normalized btn -1 and 1 (most useful for unidimentional measure of sentiment for a sentence).

Typical thresholds in the literature are as follows:

- positive sentiment: compound score >= 0.05
- neutral sentiment: compound score > -0.05 and compound score < 0.05
- negative sentiment: compound score <= -0.05

pos neu and neg scores ARE RATIOS for proportions of text that fall in each category, so they should add up to 1 (or close to it; most useful for multidimensional measures of sentiment for a sentence)


from VADER documentation:
- VADER sentiment lexicon sensitive to both polarity and intensity of sentiments expressed in social media contexts
- lexical featuers include 'Western style' emoticons (with ASCII characters); acronyms/initilaisms (lol/wtf)
- commonly used slang w/ sentiment value
* WORD-ORDER SENSITIVE RELATIONS BETWEEN TERMS (intensifiers impact sentiment intensity)

## Medium Sentiment Analysis article code

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('all')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')

df.head()
```

```python
def preprocess_text(text):
    
    """tokenize text"""
    tokens = word_tokenize(text.lower())
    
    """remove stop words"""
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]
    
    """lemmatize tokens"""
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    
    """make tokens into a string again"""
    processed_text = ' '.join(lemmatized_tokens)
    
    return processed_text
```

```python
df['preprocessedText'] = df['reviewText'].apply(preprocess_text)
```

```python
df
```

```python
sent_analyzer = SentimentIntensityAnalyzer()
```

```python
"""get the polarity scores"""
"""anything that has a 1 is pos, else neg"""

def get_sentiment(text):
    scores = sent_analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return sentiment
```

```python
df['sentiment'] = df['reviewText'].apply(get_sentiment)
df.head(10)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(df['Positive'], df['sentiment']))
```

```python
from sklearn.metrics import classification_report
print(classification_report(df['Positive'], df['sentiment']))
```

```python
def analyze_sentiment(review_id: str) -> str:
    """Returns sentiment category based on the average of all sentence compound scores."""
    text = nltk.corpus.movie_reviews.raw(review_id)
    sia = SentimentIntensityAnalyzer()
    
    scores = [
        sia.polarity_scores(sentence)["compound"]
        for sentence in nltk.sent_tokenize(text)
    ]
    
    average_score = mean(scores)
    
    if average_score < -0.05:
        return "Negative"
    elif -0.05 <= average_score <= 0.05:
        return "Neutral"
    else:
        return "Positive"
```

```python
def extract_features(text):
    features = dict()
    wordcount = 0
    compound_scores = list()
    positive_scores = list()

    for sentence in nltk.sent_tokenize(text):
        for word in nltk.word_tokenize(sentence):
            if word.lower() in top_100_positive:
                wordcount += 1
        compound_scores.append(sia.polarity_scores(sentence)["compound"])
        positive_scores.append(sia.polarity_scores(sentence)["pos"])

    # Adding 1 to the final compound score to always have positive numbers
    # since some classifiers you'll use later don't work with negative numbers.
    features["mean_compound"] = mean(compound_scores) + 1
    features["mean_positive"] = mean(positive_scores)
    features["wordcount"] = wordcount

    return features
```
