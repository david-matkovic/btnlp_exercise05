```python
def get_sentiment_label(compound_score):
    if compound_score >= 0.05:
        return "Positive"
    elif -0.05 < compound_score < 0.05:
        return "Neutral"
    else:
        return "Negative"
```

```python
def analyze_sentiment_of_words_in_sentences(sentences):
    sid = SentimentIntensityAnalyzer()
    
    for sentence in sentences:
        words = sentence.split()
        
        for word in words:
            ss = sid.polarity_scores(word)
            print(f"Sentence: {sentence}, Word: {word}, Positive: {ss['pos']}, Neutral: {ss['neu']}, Negative: {ss['neg']}")
```
